```python
#Problem No. 2

def majority_votes(lst:list) -> str:
    
    '''This function returns the majority vote in a list
    
       A majority vote is an element that occurs > N/2 times
       in a list (where N is the length of the list)'''
    
    my_set = set(lst)
    
    for i in my_set:
        if lst.count(i) > (len(lst)/2):
            return f'{i} has got the majority votes'
        
        elif lst.count(i) == len(lst)/len(my_set):
            return f'No one has got the majority votes'
```

```python
majority_votes(["A", "A", "C", "A", "B", "A"])
```

```python
#Problem No. 3

def censor_string(str_1:str,lst:list,str_2:str) -> str:
    
    '''This function takes a string txt and censors any word
       from a given list. The text removed must be replaced by
       the given character char.
       
       Example :
       censor_string("Today is a Wednesday!",["Today", "a"],"-")
       -> "----- is - Wednesday!"'''
    
    S = str_1.split()
    
    for i in lst:
        if i in S:
            S.insert(S.index(i),str_2*len(i))
            S.remove(i)

    return ' '.join(S)
```

```python
censor_string('Why did the chicken cross the road ?',['did','chicken','road'],'*')
```

```python
#Problem No. 4

def is_polydivisible(n:int) -> bool:
    
    '''A Polydivisible Number (or magic number) is a number in a
       given number base with digits abcde... that has the following
       properties :
       - Its first digit a is not 0
       - The number formed by its first two digits ab is a multiple of 2
       - The number formed by its first three digits abc is a multiple of 3
       - The number formed by its first four digits abcd is a multiple of 4
       
       This function takes an integer n and returns True if the
       given number is a Polydivisible Number and False otherwise'''
    
    S = str(n)
    
    for i in range(1,len(S)+1):
        if int(S[:i])%i == 0:
            continue
        else:
            return False
            break
    return True
```

```python
is_polydivisible(123220)
```

```python
#Problem No. 5

def sum_primes(lst:list) -> int:
    
    '''This function takes a list of numbers and returns the
       sum of all prime numbers in the list'''
    
    def check_prime(n:int) -> bool:
        
        from math import sqrt, modf
        
        frac = modf(sqrt(n))[0]
        
        if n<=3 or n == 5:
            return n>1
        
        elif n%2 == 0 or n%3 == 0:
            return False
        
        elif n%5 == 0 or frac == 0:
            return False
        
        i = 5
        while i**2 <= n:
            if n%i == 0 or n%(i+2) == 0:
                return False
            i += 6
        return True

    count = 0

    for i in lst:
        if check_prime(i):
            count += i
            
    return f'Sum of prime numbers in given list is {count}'
```

```python
sum_primes([2,3,4,11,20,50,71])
```
