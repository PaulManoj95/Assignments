```python
#Problem No. 1

def pentagonal():
    
    '''This function takes a positive integer number (N) and 
    calculates how many dots exist in a pentagonal shape around
    the center dot on the Nth iteration'''
    
    N = int(input('How many iterations ?\n'))
    count = 1
    
    for i in range(1,N+1):
        count += 5*(i - 1)
    
    print(f'Total number of dots in the whole pentagon'
          f' on the {N}th iteration is {count}')
```

```python
pentagonal()
```

```python
#Problem No. 2

def encrypts(Input:str) -> str:
    
    '''This function takes a word as input and return
    an encrypted output through some internal modifications'''
    
    Reversed_Str = Input.lower()[::-1]
    vowels = {'a':'0','e':'1','i':'2','o':'2','u':'3'}
    
    for i in Reversed_Str:
        if i in vowels.keys():
            Reversed_Str = Reversed_Str.replace(i,vowels[i])
    
    return f'Encrypted output : {Reversed_Str+"aca"}'
```

```python
encrypts('Reindeer')
```

```python
#Problem No. 3

def check_for_day(month:int,year:int) -> bool:
    
    '''This function takes month and year as integer arguments and
    day of the week and date from user and returns True or False
    based on whether the given day belongs to that date or not'''
    
    from datetime import date
    
    My_Dict = {'MONDAY':0, 'TUESDAY':1, 'WEDNESDAY':2,
              'THURSDAY':3, 'FRIDAY':4, 'SATURDAY':5, 'SUNDAY':6}
    
    Input = input('Which day of the week you want to check for ?\n').upper()
    day = int(input('Which date you want to check this day with ?\n'))
    
    return date(year,month,day).weekday() == My_Dict[Input]
```

```python
check_for_day(12,2021)
```

```python
#Problem No. 4

def count_bad_cookies(lst:list):
    
    '''This function uses RegEx Negative Lookbehind that helps
       in counting how many bad cookies are produced every day
       from a given list'''
    
    import re
    
    S = ','.join(lst)
    
    pattern = '(?<!cookie)bad'
    match = re.findall(pattern,S)
    
    print(f'No. of bad cookies produced everyday : {len(match)}')
```

```python
count_bad_cookies(["bad cookie", "good cookie", "bad cookie", 
                   "good cookie", "good cookie"])
```

```python
#Problem No. 5

def pluralize(lst:list) -> set:
    
    '''Given a list of words in the singular form,
       this function returns a set of those words in the 
       plural form if they appear more than once in the list'''
    
    S = set(lst)
    
    for i in S:
        if lst.count(i)>1:
            S.remove(i)
            S.add(i+'s')
    return S
```

```python
pluralize(['cow', 'dog', 'cow', 'cow'])
```
