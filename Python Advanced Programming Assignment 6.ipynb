```python
# Problem No. 1

def find_the_difference(s1:str,s2:str) -> str:
    
    '''Given two strings s1 and s2. String s2 is generated by
       randomly shuffling string s1 and then adding one more letter
       at a random position.
       
       This function returns the letter that was added to s1'''
    
    from itertools import permutations
    
    perm_list = permutations(s2)
    lst = []
    
    for perm in perm_list:
        lst.append(''.join(perm))
    
    if len(s1) == 0:
        return s2
        
    elif len(s1) != 0:
        for i in lst:
            if i[:-1] == s1:
                return i[-1]
```

```python
find_the_difference('abcd','axcdb')
```

```python
# Problem No. 2

def count_datatypes(*args):
    
    '''This function accepts unlimited arguments, checks and counts
       how many data types are in those arguments and finally
       returns the total in a list.
       
       List order is:
       [Int, Str, Bool, List, Tuple, Dictionary]'''
        
    Int = 0
    Str = 0
    Bool = 0
    List = 0
    Tuple = 0
    Dictionary = 0
    
    for i in args:
        if type(i) == int:
            Int += 1
        if type(i) == str:
            Str += 1
        if type(i) == bool:
            Bool += 1
        if type(i) == list:
            List += 1
        if type(i) == tuple:
            Tuple += 1
        if type(i) == dict:
            Dictionary += 1
    
    return [Int, Str, Bool, List, Tuple, Dictionary]
```

```python
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
# Problem No. 3

def fib_str(n:int,lst:list):
    
    '''A Fibonacci string is a precedence of the Fibonacci series
       This function works with any two characters of the English
       alphabet (as opposed to the numbers 0 and 1 in the Fibonacci
       series) as the initial items and concatenates them together
       as it progresses in a similar fashion as the Fibonacci series
       and '''
    
    s_1 = lst[0]
    s_2 = lst[1]
    
    count = 2
    
    while count<n:
        s_1,s_2 = s_2,s_2+s_1
        lst.append(s_2)
        count += 1
        
    return lst
```

```python
fib_str(6,['n','k'])
```

```python
# Problem No. 4

def ones_threes_nines(n1:int) -> str:
    
    '''This function returns out how many ones, threes, and nines
       it would at least take for the number of each to add up to
       the given integer when multiplied by one, three or nine'''
    
    if n1<=26 :
        
        if n1>=9:
            count_1 = n1//9
        else:
            count_1 = 0
        
        n2 = n1 - count_1*9
    
        if n2>=3:
            count_2 = n2//3
        else:
            count_2 = 0
        
        n3 = n2 - count_2*3

        if n3>=1:
            count_3 = n3//1
        else:
            count_3 = 0
            
        return f'nines:{count_1}, threes:{count_2}, ones:{count_3}'
    
    else:
        return 'Number limit exceeded !'
```

```python
ones_threes_nines(26)
```

```python
# Problem No. 5

def fastFib(n:int) -> int:
    
    '''The function takes an integer n as argument and returns
       the nth fibonacci number Fn'''
    
    f_1 = 0
    f_2 = 1
    count = 2
    
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        while count<=n:
            f_1,f_2 = f_2,f_1+f_2
            count += 1
    return f_2
```
