```python
#Problem No. 1

def fastFib(n:int) -> int:
    
    '''The function takes an integer n as argument and returns
       the nth fibonacci number Fn'''
    
    f_1 = 0
    f_2 = 1
    count = 2
    
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        while count<=n:
            f_1,f_2 = f_2,f_1+f_2
            count += 1
    return f_2
```

```python
fastFib(20)
```

```python
#Problem No. 2

def convert_to_hex(S:str) -> str:
    
    '''This function takes strings characters as ASCII and
       returns each characters hexadecimal value as a string'''
    
    lst = []
    
    for i in S:
        b = i.encode('utf-8')
        lst.append(b.hex())
    
    return ' '.join(lst)
```

```python
convert_to_hex('Marty Poppinson')
```

```python
#Problem No. 3

def uncensor(str_1:str,str_2:str) -> str:
    
    '''This function takes a censored string and a string of
       censored vowels and returns the original uncensored string'''
    
    for i in str_2:
        n = str_1.index('*')
        str_1 = str_1.replace(str_1[n],i,1)
    return str_1
```

```python
#Problem No. 4

def get_domain_name(ip_address:str) -> str:
    
    '''This function takes an IP address and returns the domain
       name using PTR DNS records'''
    
    import socket
    
    print('Fetching domain name using PTR DNS :')
    
    result = socket.gethostbyaddr(ip_address)[0]
    return result
```

```python
get_domain_name('208.80.154.232')
```

```python
#Problem No. 5

def fact_of_fact(n:int) -> int:
    
    '''This function takes an integer n and returns the factorial
       of factorials.
       
       Example : fact_of_fact(4) -> 4! * 3! * 2! * 1! = 288'''

    def fact(n):
        mul_1 = 1
        for i in range(1,n+1):
            mul_1 *= i
            
        return mul_1
        
    mul_2 = 1
    
    for j in range(1,n+1):
        mul_2 *= fact(j)
        
    return mul_2
```

```python
fact_of_fact(6)
```
